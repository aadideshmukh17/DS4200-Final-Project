```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
alt.data_transformers.disable_max_rows()
```

```python
df_2021 = pd.read_csv('Food_security_2021.csv')
df_2017 = pd.read_csv('Food_security_2017.csv')
df_2015 = pd.read_csv('Food_security_2015.csv')

def filter_df_scatter(df, year_label):
    df = df.copy()
    df = df[~df['FSD012N'].isin([77, 99])]
    df = df.dropna(subset=['FSD012N'])
    df = df[~df['FSD225'].isin([77777, 99999])]
    df = df.dropna(subset=['FSD225'])
    df['FSD225'] = df['FSD225'].replace(55555, 300)
    df = df[['FSD225', 'FSD012N']].copy()
    df['Year'] = year_label
    df = df.replace(5.397605346934028e-79, 0)
    return df


df_2015_scatter = filter_df_scatter(df_2015, '2015-2016')
df_2017_scatter = filter_df_scatter(df_2017, '2017-2018')
df_2021_scatter = filter_df_scatter(df_2021, '2021-2023')

df_combined_scatter = pd.concat([df_2015_scatter, df_2017_scatter, df_2021_scatter], ignore_index=True)
# df_combined = df_combined.sample(n=2000, random_state=42)

selection = alt.selection_point(fields=['Year'], bind='legend')

scatterplot = alt.Chart(df_combined_scatter).mark_point().encode(
    x=alt.X('FSD225:Q', title='Number of Days'),
    y=alt.Y('FSD012N:Q', title='Number of People Receiving Benefits'),
    color=alt.condition(
        selection,
        alt.Color('Year:N', title='Year (select one)', scale=alt.Scale(
            domain=['2015-2016', '2017-2018', '2021-2023'],
            range=['#27A6F5', '#C4122D', '#5FEF0C']
        )),
        alt.value('transparent')
    ),
    tooltip=[
        alt.Tooltip('Year:N', title='Time Period'),
        alt.Tooltip('FSD225:Q', title='Number of Days'),
        alt.Tooltip('FSD012N:Q', title='Number of People Receiving Benefits')
    ]
).add_params(
    selection
).properties(
    title='Number of Recipients and Time Since Last Receiving Benefits',
    width=650,
    height=500
).configure_title(
    fontSize=20,  # Set the desired font size
    font='Times New Roman',  # Optionally, set the font family
    anchor='middle',  # Anchor the title to the start
    color='black'  # Set the title color
).configure_axis(
    titleFontSize=18,
    titleFont='Times New Roman'
)

scatterplot.save('Aadi-scatterplot.html')
scatterplot
```

```python
def filter_df_boxplot(df, year_label):
    df = df.copy()
    df = df[~df['FSD670ZC'].isin([777, 999])]
    df = df.dropna(subset=['FSD670ZC'])
    df['Year'] = year_label
    df = df.replace(5.397605346934028e-79, 0)
    return df

df_2015_boxplot = filter_df_boxplot(df_2015, '2015-2016')
df_2017_boxplot = filter_df_boxplot(df_2017, '2017-2018')
df_2021_boxplot = filter_df_boxplot(df_2021, '2021-2023')

df_combined_boxplot = pd.concat([df_2015_boxplot, df_2017_boxplot, df_2021_boxplot], ignore_index=True)

boxplot = alt.Chart(df_combined_boxplot).mark_boxplot(extent='min-max').encode(
    x=alt.X('FSD670ZC:Q', title='Number of Months'),
    y=alt.Y('Year:N', title='Year'),
    color=alt.Color('Year:N', scale=alt.Scale(
        domain=['2015-2016', '2017-2018', '2021-2023'],
        range=['#27A6F5', '#C4122D', '#5FEF0C']
    ))
).properties(
    title='Number of Months Receiving WIC Benefits',
    width=650,
    height=500
).configure_title(
    fontSize=20,
    font='Times New Roman',
    anchor='middle',
    color='black'
).configure_axis(
    titleFontSize=18,
    titleFont='Times New Roman'
)

boxplot.save('Aadi-boxplot.html')
boxplot
```
